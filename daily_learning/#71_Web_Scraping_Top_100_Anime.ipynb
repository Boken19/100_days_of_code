```python
#importing useful libraries and classes
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
```

```python
# html upload
my_url = "https://www.imdb.com/list/ls057577566/"
uClient = uReq(my_url)
page_html = uClient.read()
uClient.close()
```

```python
#html parser
page_soup = soup(page_html, 'html.parser')
S = soup(uClient)
```

```python
page_soup
```

```python
containers = page_soup.findAll('div', {'class': 'lister-item mode-detail'})
print(len(containers))
```

```python
print(soup.prettify(containers[0]))
```

```python
# extract data into a csv file 
filename = 'imdb_anime.csv'
f = open(filename, 'w')
headers = 'Name, Year, Runtime, Rating, Votes\n'
f.write(headers)

for container in containers:
    name_a = container.findAll('a')
    name = name_a[1].text.replace(',',"")
    year_a = container.findAll('span', {'class': 'lister-item-year'})
    year = year_a[0].text
    runtime_a = container.findAll('span', {'class': 'runtime'})
    if runtime_a:
        runtime = runtime_a[0].text
    else:
        runtime = 'NULL'
    rating_a = container.findAll('span', {'class': 'ipl-rating-star__rating'})
    rating = rating_a[0].text
    votes_a = container.findAll('span', {'name': 'nv'})
    votes = votes_a[0].text.replace(',',"")
    print(name + ',' + year + ',' + runtime + ',' + rating + ',' + votes + '\n')
    f.write(name + ',' + year + ',' + runtime + ',' + rating + ',' + votes + '\n')
```
