```python
# String Manipulation : Intermediate Level

#Greatest="Albert Einstein,Marie Curie,Issac Newton"

# Expected Output:

# Einstein.A
# Curie.M
# Newton.I
```

```python
greatest="Albert Einstein,Marie Curie,Issac Newton"
l1 = greatest.split(',')
l2 = []
for char in l1:
    l2.append(char.split())
for i in l2:
    print(i[1]+'.'+i[0][0])
```

```python
## Guess the Number | Decision Making | Looping | R & D

# 2 Players

# computerChoice : generate a random number between 1-20
# playerChoice : ask the user to enter a number between 1-20

# if the playerChoice is greater then computerChoice:
    # display "your guess is too high"
    
# if the playerChoice is lesser then computerChoice:
    # display your guess is too low
    
# if the playerChoice and  computerChoice matches:
    # display: "Gotcha! you guessed  it right"
    
#please note: a user can attempt a max of 6 guesses after 6 guesses
    # display "a oh ! you have exhausted all your attempts"
```

```python
from random import randint

random_no = randint(1, 21)
playerChoice = 0
for i in range(0,6):
    playerChoice = int(input("Guess a random no. between 1 and 20: "))
    if playerChoice > random_no:
        print("Your guess is too high.")
    elif playerChoice < random_no:
        print("Your guess is too low.")
    else:
        print('Gotcha! you guessed it right')
        break
else:
    print('Try later')
    
        
```

```python
#create a new list 'pokemonTitles' and extract only the pokemon names
pokemonTitles = []
for i in pokemons:
    a = i.split(',')
    pokemonTitles.append(a[0])
pokemonTitles
```

```python
# create a new list of distinct(unique) pokemonTypes

pokemonTypes = []
for i in pokemons:
    b = i.split(',')
    pokemonTypes.append(b[1].rstrip('\n'))
set(pokemonTypes)
```
