```python
df.drop(columns = ['SNo', 'Country', 'RegionCode'], inplace = True)
```

```python
import re
```

```python
df['month'] = df.Date.apply(lambda x: re.split('[-T]',x)[1])
```

```python
df['day'] = df.Date.apply(lambda x: re.split('[-T]',x)[2])
```

```python
df.head()
```

```python
df.RegionName.unique()
```

```python
df.groupby(by = 'RegionName').sum().loc[:,['TotalPositiveCases','Deaths', 'Recovered']].plot(figsize = (15,6))
plt.xticks(rotation = 'vertical')
```

```python
# Region wise total positive cases over the year 2020
df.groupby(by = 'RegionName').sum()['TotalPositiveCases'].plot.bar(figsize = (10,6))
plt.xlabel('Region', fontdict = {'fontsize' : 14, 'fontweight': 5, 'color': 'red'})
plt.ylabel('Total Positive Cases', fontdict = {'fontsize' : 14, 'fontweight': 5, 'color': 'red'})
```

- Lombardia is the region with most covid cases because it consists of more than 1/6th of Italy population

```python
# month wise covid cases 
plt.subplot(1,2,1)
df.groupby(['month']).sum()['TotalPositiveCases'].plot(figsize = (15,6))
df.groupby(['month']).sum()['Recovered'].plot()
df.groupby(['month']).sum()['Deaths'].plot()
plt.xlabel('Month', fontdict = {'fontsize' : 14, 'fontweight': 5, 'color': 'red'})
plt.ylabel('No. of cases', fontdict = {'fontsize' : 14, 'fontweight': 5, 'color': 'red'})
plt.legend()
plt.subplot(1,2,2)
df.groupby(['month']).sum()['Deaths'].plot(figsize = (15,6), c = 'g')
plt.xlabel('Month', fontdict = {'fontsize' : 14, 'fontweight': 5, 'color': 'red'})
plt.ylabel('No. of cases', fontdict = {'fontsize' : 14, 'fontweight': 5, 'color': 'red'})
plt.legend()
plt.show()
```

- We can see from above graph that in the months of September and October the Covid cases started to rise rapidly and in November the cases reached its peak and started declining afterwards.
- We can clearly see that when the cases were highest the death cases were also heighest.
