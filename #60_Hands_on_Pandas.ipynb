### Filter the data based on conditions

```python
# Filter the rows where gender is male
df['gender'] == 'Male'
```

```python
df[df['gender'] == 'Male'].head()
```

```python
df[df['InternetService'] == 'DSL'].head()
```

```python
# More than one Condition
# filter the rows where gender is female and have monthly charges> 20
df[(df['gender'] == 'Female') &  (df['MonthlyCharges'] > 20)].head()
```

```python
# find out the number of customers who are seniorcitizen and hold month-to-month contract
df[(df['SeniorCitizen'] == True) & (df['Contract'] == 'Month-to-month')].shape
```

```python
df.loc[(df['SeniorCitizen'] == True) & (df['Contract'] == 'Month-to-month'),:]
```

### Sort

```python
# By default the data will be sorted in ascending order
# ascending by default True
df.sort_values( by = 'tenure')
```

```python
df.sort_values( by = 'tenure', ascending = False)
```

```python
# Sort the data in descending order for Monthly charges and tell me the customerID of the first 5
df.sort_values( by = 'MonthlyCharges', ascending = False).head()
```

#### How to perform sorting on more than one columns?

```python
df.sort_values( by = ['tenure', 'MonthlyCharges'], ascending = [False, True])
```

### Grouping Data

```python
df.groupby('gender').agg({'tenure': 'mean', 'customerID': 'count'})
```

```python
df.groupby('gender').sum()
```

```python
df.groupby('gender')['tenure'].mean()
```
